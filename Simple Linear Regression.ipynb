```python
import numpy as np
import matplotlib.pyplot as plt
```

Generate training data.

```python
x = np.arange(0, 10, 0.1)
y = 1 + (2*x) + (np.random.normal(0, 1, len(x)) * 5)
```

Learn the simple linear regression coefficients from data.

```python
mx = x.mean()
my = y.mean()
temp = x - mx
c1 = np.sum(temp * (y - my)) / np.sum(temp ** 2)
c0 = my - (c1 * mx)
```

Plot result.

```python
x2 = [0, 10]
y2 = [c0 + c1*0, c0 + c1*10]

my_dpi = 96
plt.figure(figsize=(600/my_dpi, 600/my_dpi), dpi=my_dpi)

plt.scatter(x, y, color='b', s=30)
plt.plot(x2, y2, color='r', linewidth=3)

plt.axis([0,10,-5,30])

plt.xlabel('X', fontsize=12)
plt.ylabel('Y', fontsize=12)
plt.title('Simple linear regression\n', fontsize=12)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.show()
```
